## 1. Point Operation

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

img = cv.imread('./data/lena.jpg', 0)
img1 = np.copy(img)
img2 = np.copy(img)
img3 = np.copy(img)

for y in range(img1.shape[0]):
    for x in range(img1.shape[1]):
        img1[y,x] = min(img1[y,x]+100, 255) #밝게

for y in range(img2.shape[0]):
    for x in range(img2.shape[1]):
        img2[y,x] = max(img2[y,x]-100, 0) #어둡게

for y in range(img3.shape[0]):
    for x in range(img3.shape[1]):
        img3[y,x] = 255 - img3[y,x] #반전
        
plt.subplot(2,2,1)
plt.title('original')
plt.axis('off')
plt.imshow(img, cmap = 'gray')

plt.subplot(2,2,2)
plt.title('bright')
plt.axis('off')
plt.imshow(img1, cmap = 'gray')

plt.subplot(2,2,3)
plt.title('dark')
plt.axis('off')
plt.imshow(img2, cmap = 'gray')

plt.subplot(2,2,4)
plt.title('reverse')
plt.axis('off')
plt.imshow(img3, cmap = 'gray')
plt.show()
```

```python
img4 = np.copy(img)
gamma = [0.1, 1, 2.5]

for i,g in enumerate(gamma):
    for y in range(img4.shape[0]):
        for x in range(img4.shape[1]):
            img4[y,x] = 255 * ((img4[y,x]/255)**g) #감마 수정
    
    plt.subplot(1,3,i+1)
    plt.title(g)
    plt.axis('off')
    plt.imshow(img4, cmap = 'gray')
    
    img4 = np.copy(img) #초기화
    
plt.show()
```

## 3. Geometric Operation

```python
import cv2
src = cv2.imread('./data/lena.jpg')

rows, cols, channels = src.shape
M1 = cv2.getRotationMatrix2D( (rows/2, cols/2),  45, 0.5 )
M2 = cv2.getRotationMatrix2D( (rows/2, cols/2), -45, 1.0 )

dst1 = cv2.warpAffine( src, M1, (rows, cols))
dst2 = cv2.warpAffine( src, M2, (rows, cols))

cv2.imshow('dst1',  src)
cv2.imshow('dst2',  dst2)
cv2.waitKey()    
cv2.destroyAllWindows()
```
